```python
from pathlib import Path
import csv
csvpath = Path("budget_data.csv")

sum_of_pnl = 0
months = []
pnl = []
diff = []
new_list = {}


with open(csvpath, "r") as csvfile:
    
    csvreader = csv.reader(csvfile, delimiter = ",")
    csv_header = next(csvreader)
    
    for row in csvreader:
       
        months.append(row[0])
        total_months = len(months)
        pnl_row = int(row[1])
        pnl.append(pnl_row)
        sum_of_pnl = sum(pnl)
        
    for x in range(1, len(pnl)):
        result = (pnl[x]-pnl[x-1])
        diff.append(result)
        avg = sum(diff) / (total_months - 1)
        avg = round(avg, 2)
            
        max_diff = max(diff)
        min_diff = min(diff)
        if pnl[x] - max_diff == pnl[x-1]:
            max_date_value = pnl[x]
        if pnl[x] - min_diff == pnl[x-1]:
            min_date_value = pnl[x]
          

            
new_list = dict(zip(months, pnl))

for months, pnl in new_list.items():
    if pnl == max_date_value:
        max_month = months
        
for months, pnl in new_list.items():
    if pnl == min_date_value:
        min_month = months
    
                                    



print ("Financial Analysis")
print ("------------------------------------------")
print (f"The total number of months is {total_months}.")
print (f"The sum of profit and loss is ${sum_of_pnl}.")
print (f"The average change in profit/loss per month is ${avg}.")
print (f"Greatest Increase in Profits is ${max_diff} on {max_month}.")
print (f"Greatest Decrease in Profits is ${min_diff} on {min_month}.")


with open(Path("financial_analysis.txt"), "x") as outfile:
    outfile.write(str("Financial Analysis"))
    outfile.write(str("------------------------------------------"))
    outfile.write(str(f"The total number of months is {total_months}."))
    outfile.write(str(f"The sum of profit and loss is ${sum_of_pnl}."))
    outfile.write(str(f"The average change in profit/loss per month is ${avg}."))
    outfile.write(str(f"Greatest Increase in Profits is ${max_diff} on {max_month}."))
    outfile.write(str(f"Greatest Decrease in Profits is ${min_diff} on {min_month}."))
    
    
      
           



```
